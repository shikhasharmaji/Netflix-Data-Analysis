```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('clean_netflix_data.csv')
```

```python
sns.set_style('whitegrid')
```

# 1. What is the most frequent genre of movies released on Netflix

```python
df['Genre'].describe()
```

```python

# Set an aesthetic style and color palette
sns.set_theme(style="whitegrid")
sns.set_palette("Reds_r")

# Create the count plot
plot = sns.catplot(
    y='Genre',
    data=df,
    kind='count',
    order=df['Genre'].value_counts().index,
    height=6,       # plot height
    aspect=1.5,     # width/height ratio
    palette='Reds_r'
)

# Add title and labels
plot.set_titles("")
plot.set_axis_labels("Count", "Genre")
plt.title(" Genre Distribution in Movies", fontsize=16, fontweight='bold', pad=20)

# Make the grid and spines clean
sns.despine(left=True, bottom=True)

# Add value labels beside bars
for container in plot.ax.containers:
    plot.ax.bar_label(container, fmt='%d', label_type='edge', fontsize=10, padding=3)

# Adjust layout for neatness
plt.tight_layout()
plt.show()
plt.savefig('most_frequent_genre_of_movies.png')


```

# 2. Which has highest votes in vote average column

```python
df.head()
```

```python

# Set theme and color palette
sns.set_theme(style="whitegrid")
sns.set_palette("Reds_r")

# Create the plot
plot = sns.catplot(
    y='Vote_Average',
    data=df,
    kind='count',
    order=df['Vote_Average'].value_counts().index,
    height=6,
    aspect=1.4,
    color='red'
)

# Remove axis titles (for cleaner look)
plot.set_titles("")
plot.set_axis_labels("Count", "Vote Average")

# Add a bold title
plt.title(" Vote Average Distribution", fontsize=16, fontweight='bold', pad=20)

# Clean borders and grid
sns.despine(left=True, bottom=True)

# Add count labels beside bars
for container in plot.ax.containers:
    plot.ax.bar_label(container, fmt='%d', label_type='edge', fontsize=10, padding=3)

# Rotate y-axis labels if they overlap
plot.ax.set_yticklabels(plot.ax.get_yticklabels(), rotation=0, fontsize=10)

# Tight layout for perfect spacing
plt.tight_layout()

# Show the plot
plt.show()
```
